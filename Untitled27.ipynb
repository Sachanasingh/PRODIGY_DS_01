```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load the data
data = pd.read_csv('Cerebral Stroke Prediction.csv')

# Set the visual style for the plots
sns.set(style="whitegrid")

# Create a figure with two subplots: one for the histogram and one for the bar chart
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histogram for age distribution
sns.histplot(data['age'], bins=20, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

# Bar chart for gender distribution
sns.countplot(x='gender', data=data, palette='Set2', ax=axes[1])
axes[1].set_title('Gender Distribution')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
